```python
import numpy as np
x= np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,y,color='red')
ax1.set_ylabel('Y')



ax2=fig.add_axes([0.2,0.60,0.3,0.3])
ax2.plot(x,y,color='red')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,z)

ax1.set_xlabel('X')
ax1.set_ylabel('Z')





ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='Blue',lw=3,ls='--')

axes[1].plot(x,z,color='Red',lw=3)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,color='Blue',lw=3,ls='-')

axes[1].plot(x,z,color='Red',lw=3,ls='--')
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='-',marker='o',markersize=10,markerfacecolor='yellow',
       markeredgewidth=3,markeredgecolor='green')


plt.show()
```

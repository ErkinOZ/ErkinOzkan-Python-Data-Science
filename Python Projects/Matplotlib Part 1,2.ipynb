```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y,'r--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')

plt.subplot(1,2,2)
plt.plot(y,x,'r--')
```

```python
#ObjectOrientedMethod

fig = plt.figure()

axes = fig.add_axes([0.7,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.45,0.3,0.4])

axes1.plot(x,y)
axes1.set_title('Big PLot')

axes2.plot(y,x,'r--')
axes2.set_title('Small')
plt.show()
```

```python
fig=plt.figure()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'r--')
axes[0].set_title('0 Titel')

axes[1].plot(y,x)
axes[1].set_title('1 Title')

plt.tight_layout()
```
